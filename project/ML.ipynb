```python
import pandas as pd
from sklearn import svm
from sklearn.model_selection import GridSearchCV
import os
import matplotlib.pyplot as plt
from skimage.transform import resize
from skimage.io import imread
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
import pickle
```

```python
Categories=['empty','occupied']

flat_data_arr=[]
target_arr=[]
#please use datadir='/content' if the files are upload on to google collab
#else mount the drive and give path of the parent-folder containing all category images folders.
datadir='./data/processed/'
for i in Categories:
    print(f'loading... category : {i}')
    path=os.path.join(datadir,i)
    for img in os.listdir(path):
        img_array=imread(os.path.join(path,img))
        img_resized=resize(img_array,(80, 80, 3))
        flat_data_arr.append(img_resized.flatten())
        target_arr.append(Categories.index(i))
    print(f'loaded category:{i} successfully')
flat_data=np.array(flat_data_arr)
target=np.array(target_arr)
df=pd.DataFrame(flat_data)
df['Target']=target
df
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=67,stratify=y)
print('Splitted Successfully')
```

```python
# param_grid={'C':[0.1,1,10,100],'gamma':[0.0001,0.001,0.1,1],'kernel':['rbf','poly']}
param_grid={'C':[0.1,1,10,100],'gamma':[0.0001,0.001,0.1,1],'kernel':['rbf']}
svc=svm.SVC(probability=True)
print("The training of the model is started, please wait for while as it may take few minutes to complete")
model=GridSearchCV(svc, param_grid, n_jobs=-1)
model.fit(x_train,y_train)
print('The Model is trained well with the given images')
model.best_params_
```

```python
y_pred=model.predict(x_test)
print("The predicted Data is :")
y_pred
```

```python
print("The actual data is:")
np.array(y_test)
```

```python

#classification_report(y_pred,y_test)
print(f"The model is {accuracy_score(y_pred,y_test)*100}% accurate")
#confusion_matrix(y_pred,y_test)
```

```python
tp, tn, fn, fp = 0, 0, 0, 0
y_test = np.array(y_test)

for i in range(len(y_test)):
    if y_test[i] == 0 and y_pred[i] == 0:
        tn += 1
    elif y_test[i] == 1 and y_pred[i] == 0:
        fn += 1
    elif y_test[i] == 0 and y_pred[i] == 1:
        fp += 1
    elif y_test[i] == 1 and y_pred[i] == 1:
        tp += 1

recall = tp / (tp + fn)
precision = tp / (tp + fp)
f1_score = 2 * precision * recall / (precision + recall)
recall, precision, f1_score
```
